```python
for row in df['attacker_commander']:
    # if the cell is a string,
    if type(row) == str:
        # split up the names by the comma, 
        #and attach it to attacker commander list,
       attacker_list.append(row.split(','))
    # otherwise,     
    else:
        # do nothing
        continue
```

```python
# Create a list of attacking commanders
attacking_commanders = []

# For each row in df['attacker_commander'],if the cell is a string, split up the names by the comma, 
# and attach it to attacker commander list, otherwise, do nothing
for item in df['attacker_commander']:
    if type(item) == str:
        attacking_commanders.append(item.split(','))
    else:
        continue 
    
# Create a list of defending commanders
defending_commanders = []        
# For each row in df.defender_commander,if the cell is a string, split up the names by the comma, 
#and attach it to defender commander list,otherwise do nothing
for item in df['defender_commander']:
    if type(item) == str:
        attacking_commanders.append(item.split(','))
    else:
        continue 
```

```python
# Create a list of commanders
commanders = []
# For each list in attacker_list, and each list in that list, strip of any leading blank space 
        # and append it to the commander list
for item in attacking_commanders:
    for element in item:
        commanders.append(element.lstrip(" "))

# For each list in defender_list,and each list in that list strip of any leading blank space 
         #and append it to the commander list
for item in defending_commanders:
    for element in item:
        commanders.append(element.lstrip(" "))   
```

```python
# Convert the commander list into a 
# set of unique names, and convert it to a list
commander_unique = list(set(commanders))
# Display the total number of commanders in the list
print(len(commander_unique))
```

```python
#  Count the number of times a commander successfully attacked

# Create a list
wins = []
# for each commander: create a score variable, then go through each cell of attacker commander where the attacker wins:
    # if the cell is a string and if the commander is in the cell
    # add one to score if not add 0 to score if the cell is not a string do nothing
    # append the score to the list
for item in commander_unique:
    score = 0
    for row in df['attacker_commander'][df['attacker_outcome']== 'win']:
        if type(row) == str:
            if item in row:
                score += 1
            else:
                score += 0
    wins.append(score)
print(len(wins))
```

```python
#  Count the number of times a commander unsuccessfully attacked

# Create a list


# for each commander:
    # create a score variable
    # go through each cell of attacker commander where the attacker loses:
        # if the cell is a string
            # and if the commander is in the cell
                # add one to score
            # if not
                # add 0 to score
        # if the cell is not a string        
            # do nothing
            
    # append the score to the list
losses = [] 
for item in commander_unique:
    score = 0
    for row in df['attacker_commander'][df['attacker_outcome']== 'loss']:
        if type(row) == str:
            if item in row:
                score += 1
            else:
                score += 0
    losses.append(score)
print(len(wins))     
```

```python
#  Count the number of times a commander successfully defended
defender_wins = []
for item in commander_unique:
    score = 0
    for row in df['defender_commander'][df['attacker_outcome']== 'loss']:
        if type(row) == str:
            if item in row:
                score += 1
            else:
                score += 0
    defender_wins.append(score)
print(len(defender_wins))
```

```python
#  Count the number of times a commander unsuccessfully defended

defender_loss = []
for item in commander_unique:
    score = 0
    for row in df['defender_commander'][df['attacker_outcome'] == 'win']:
        if type(row) == str:
            if item in row:
                score += 1
            else:
                score += 0
    defender_loss.append(score)
print(len(defender_loss))
```

```python
# Create a dictionary of the four score lists
score_dict = {
    'attacker_wins': wins,
    'attacker_loss': losses,
    'defender_wins': defender_wins,
    'defender_losses': defender_loss
}

# Create a dataframe from that dictionary, indexed by a commander's name
pd.DataFrame(score_dict, index=commander_unique)
```
